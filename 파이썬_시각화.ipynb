```
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline

```

**점선 그래프 그리기 **

```
data=np.random.randn(50).cumsum()
data
plt.plot(data)
plt.show()
```

**여러 그래프 그리기 **

```
plt.subplot(1,2,1)
plt.subplot(1,2,2)
plt.show()
```

```
hist_data=np.random.randn(100)
scat_data=np.arange(30)
plt.subplot(2,2,1)
plt.plot(data)
plt.subplot(2,2,2)
plt.hist(hist_data,bins=20)
plt.subplot(2,2,3)
plt.scatter(scat_data,np.arange(30)+3*np.random.randn(30))
```

**그래프 선옵션**

```
plt.plot(data,'k+')
```

**그래프 사이즈 조절** 

```
plt.figure(figsize=(10,5))
plt.plot(data,'k--')
plt.show()
```

# **그래프 겹치시 + legend 달기 + 이름달기**

```
plt.figure(figsize=(10,5))
data=np.random.randn(30).cumsum()
plt.plot(data,'k--',label='Default')
plt.plot(data,'k-',drawstyle='steps-post',label='steps-post')
plt.legend('data analysis')
plt.show()
```

```
plt.plot(np.random.randn(1000).cumsum(),'k--',label="nct")
plt.plot(np.random.randn(1000).cumsum(),'gs',label="exo")
plt.plot(np.random.randn(1000).cumsum(),'rv',label="snsd")
plt.legend()
plt.show()
```
